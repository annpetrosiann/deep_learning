```python
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import torch.nn.functional as F
from torch.optim import Adam
```

**Data Preparation and Preprocessing**

a. Load the Iris dataset

```python
# Load the Iris dataset
iris = load_iris()
X, y = iris.data, iris.target
```

```python
y = np.where(y == 1, 1, 0)
```

b. Preprocess the data (standardization) and split it into training
and testing sets.

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=55)
```

dasakan scale em arel aystex

**Implement an ANN model using PyTorch with the following
architecture**

```python
num_features = X.shape[1]
print("Number of features:", num_features)
```

qani vor input layeri neuron-nery petq e havasar linen feauture-neri qanakin, shape-i mijocov qanaky chstel em u da vercrel vorpes input neurons qanak

```python
class Net(nn.Module):
  def __init__(self):
    super(Net,self).__init__()
    self.fc1 = nn.Linear(4,10)
    self.fc2 = nn.Linear(10,5)
    self.fc3 = nn.Linear(5,1)


  def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = torch.relu(self.fc2(x))
        x = torch.sigmoid(self.fc3(x))
        return x
```

Stex arden input layer, hidden layers ev output layern em sarqel yst hamapatasxan pahanji. Skzbnakan layernerum  ogtagorcelem relu activation functiony, sakayn output layerum voroshel em sigmoid activation function ogtagorcel qani vor ayn aveli harmar e binary data-i het asxhatelu hamar. Petq e nkatel vor amen layeri output-y myus layer-i inputn e

**Training and evaluation**

a. Train the model for 10 epochs using binary cross-entropy
loss and your desired optimizer.

```python
model = Net()
```

```python
optimizer = Adam(model.parameters(),lr=3e-4)
criterion = nn.BCELoss()
```

Voroshelem vorpes optimizer ogtagorcel Adamy, shat poqr learning rate-ov. Vorpes Loss ogtagorcelem BCE lossy, qani vor ayn amenataracvac loss-n e binary data-i hamar.

Convert the data to PyTorch tensors

```python
X_train_tensor = torch.tensor(X_train, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train.reshape(-1, 1), dtype=torch.float32)
```

qani vor scale em arel u darcrel numpy, convert em anum tensori, nuyn gorcoxutyuny nerqevum arel em nayev test data-i hamar

```python
num_epochs = 10
for epoch in range(num_epochs):

    # Forward pass
    outputs = model(X_train_tensor)

    # loss
    loss = criterion(outputs, y_train_tensor)

    # Backward pass
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    print(f'Epoch {epoch+1}, Loss: {loss.item():.4f}')
```

epochneri qanaky trvac er 10, hamapatasxanabar models forward em arel, loss em hasvel, backward em arel. Backwardi jamanak  ogtagorcel em **optimizer.zero_grad() ** vor weightery accumulate chlini

b. Evaluate the model on the test set and compute the
accuracy.

Convert the test data to PyTorch tensors

```python
X_test_tensor = torch.tensor(X_test, dtype=torch.float32)
y_test_tensor = torch.tensor(y_test.reshape(-1, 1), dtype=torch.float32)
```

```python
model.eval()

total = 0
correct = 0

with torch.no_grad():  # disables gradient calculation
    outputs = model(X_test_tensor)
    _, predicted = torch.max(outputs, 1)
    total += y_test_tensor.size(0)
    correct += (predicted == y_test_tensor.view_as(predicted)).sum().item()

print(f'The testing set accuracy of the network is: {100 * correct / total}')
```

Sigmoidy talis e havanakanutyunner, orinak 0=0.4,1=0.6 u dra maximumov vorooshum e classy, aydpes unenum enq predictionnery, hamematum enq irakan y-i het ev hasvum accuracy.
